```python
import pandas as pd

def load_data(path):
    df = pd.read_csv(path)
    return df

def check_missing_values(df):
    return df.isnull().sum()

def suggest_cleaning(df):
    for column in df.columns:
        missing = df[column].isnull().sum()
        total = len(df)
        missing_percent = (missing / total) * 100

        if missing > 0:
            print("\n==============================")
            print("Column:", column)
            print("Missing values:", missing)
            print("Missing %:", round(missing_percent, 2))

            if missing_percent > 50:
                print("Suggestion: DROP column")
            elif df[column].dtype == "object":
                print("Suggestion: Fill with MODE")
            else:
                print("Suggestion: Fill with MEDIAN")

#here it is used to Run everything
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = load_data(url)

print("Missing Values:\n", check_missing_values(df))

print("\nCleaning Suggestions:")
suggest_cleaning(df)
```

```python
def auto_clean_data(df):

    df = df.copy()

    for column in df.columns:
        missing = df[column].isnull().sum()
        total = len(df)
        missing_percent = (missing / total) * 100

        if missing > 0:

            if missing_percent > 50:
                print(f"Dropping column: {column}")
                df.drop(columns=[column], inplace=True)

            elif df[column].dtype == "object":
                print(f"Filling {column} with MODE")
                df[column] = df[column].fillna(df[column].mode()[0])

            else:
                print(f"Filling {column} with MEDIAN")
                df[column] = df[column].fillna(df[column].median())

    return df
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = load_data(url)

cleaned_df = auto_clean_data(df)

cleaned_df.isnull().sum()
```
